As we mentioned above , you can also convert the numbers into words. This could be done by using the inflect library

```python
#import the library

import inflect
q=inflect.engine()

#convert number into text
def convert_num(text):
  #splits string inot list of texts
  temp_string=text.split()
  #initialise empty list
  new_str=[]

  for word in temp_string:
    #if text is a digit , convert the digit
    #to numbers and append into the new_str list
    if word.isdigit():
      temp=q.number_to_words(word)
      new_str.append(temp)

   #append the texts as it is
    else:
      new_str.append(word)

#join the texts of new_tr to form a string
  temp_str=' '.join(new_str)
  return temp_str

input_str="you bought 6 candies form shop and 4 candies are in home"
convert_num(input_str)
```

```python
import nltk
nltk.download('punkt')
from collections import Counter

#step 1: Define the dataset(sentences)
sentences=['cat eats fish','dog eats fish','dog chases cat']

#step2: Tokinise the sentences
tokenized_sentences=[nltk.word_tokenize(sentence.lower()) for sentence in sentences]

#step3: Create vocabulary
#flatten the lsit of tokenised sentences to get all words
all_words=[word for sentence in tokenized_sentences for word in sentence]

#Get unique words(vocubulary) using counter or set
vocabulary=list(Counter(all_words).keys())

#step4 : Create a function for one-hot encoding
def one_hot_encoded(word,vocabulary):
  vector=[0]*len(vocabulary ) #Initialize vector of all 0s
  index=vocabulary.index(word) #find the index of the word in vocabulary
  vector[index]=1              #set the corresponding index to 1
  return vector

#step5 : apply one-hot encoding to each sentence
encoded_sentences=[]
for sentence in tokenized_sentences:
  encoded_sentence=[one_hot_encoded(word,vocabulary) for word in sentence]
  encoded_sentences.append(encoded_sentence)

#Output the results
print("Vocabulary: ",vocabulary)
for i,sentence in enumerate(encoded_sentences):
  print(f"sentence {i+1}:")
  for word , vector in zip(tokenized_sentences[i],sentence):
    print(f" {word}:{vector}")

```

```python

def one_hot_encode(word,vocabulary):
  vector=[0]*len(vocabulary) #[0,0,0,0,0]
  index=list(vocabulary).index(word)
  vector[index]=1
  return vector
```

```python
encoded_sentences=[]
for sentence in tokenized_sentences:
  encoded_sentence=[one_hot_encode(word,vocabulary) for word in sentence]
  encoded_sentences.append(encoded_sentence)
# print(encoded_sentences)
print(tokenized_sentences)
for x in encoded_sentences:
  print(x)
```
