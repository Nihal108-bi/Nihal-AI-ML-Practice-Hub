```python
import pandas as pd
messages=pd.read_csv('/smsspam.txt',sep='\t',names=['label','message'])
```

```python
messages
```

```python
#data cleaning and preprocessing

import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
len(messages)
```

```python
corpus=[]
for i in range(0,len(messages)):
  review=re.sub('[^a-zA-Z]',' ',messages['message'][i])
  review=review.lower()
  review=review.split()
  review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
  review=' '.join(review)
  corpus.append(review)
```

```python
corpus
```

```python
#create a bag of words
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2500)
X=cv.fit_transform(corpus).toarray()
```

```python
X
```

```python
import numpy as np
np.set_printoptions(edgeitems=30,linewidth=10000,formatter=dict(float=lambda x:"%.3g"%x))
X
```

```python
cv.vocabulary_
```

```python
#create a bag of words  #bigrams
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2500,binary=True,ngram_range=(2,3))
X=cv.fit_transform(corpus).toarray()
```

TF_IDF Practical

```python
import pandas as pd
messages=pd.read_csv('/smsspam.txt',sep='\t',names=['label','message'])
```

```python
### data clening and preprocessing
import re
import nltk
nltk.download('wordnet')
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
wordlemmatize=WordNetLemmatizer()
```

```python
corpus=[]
for i in range(0,len(messages)):
  review=re.sub('[^a-zA-Z]',' ',messages['message'][i])
  review=review.lower()
  review=review.split()
  review=[wordlemmatize.lemmatize(word) for word in review if not word in stopwords.words('english')]
  review=' '.join(review)
  corpus.append(review)
```

```python
corpus
```

**Create TF-IDF and NGrams**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer(max_features=100)
X=tfidf.fit_transform(corpus).toarray()
```

```python
X
```

```python
import numpy as np
np.set_printoptions(edgeitems=30,linewidth=10000,formatter=dict(float=lambda x:"%.3g"%x))
X
```

N Grams

```python
tfidf=TfidfVectorizer(max_features=100,ngram_range=(2,2))
X=tfidf.fit_transform(corpus).toarray()
tfidf.vocabulary_
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense,Embedding,Input
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
import numpy as np
from keras.models import Sequential
from keras.layers import SimpleRNN, Dense

def simple_sequence_prediction():
    """
    Predict the next number in a sample sequence.
    """
    # Generate sample sequence
    X = np.array([[i + j for j in range(3)] for i in range(100)])  # Input sequences
    y = np.array([i + 3 for i in range(100)])                     # Next number in sequence

    # Reshape X to 3D for RNN input (samples, timesteps, features)
    X = X.reshape((100, 3, 1))

    # Build the model
    model = Sequential([
        SimpleRNN(32, input_shape=(3, 1)),  # RNN layer
        Dense(1)                            # Output layer
    ])
    model.compile(optimizer='adam', loss='mse')

    # Train the model
    model.fit(X, y, epochs=100, verbose=1)

    # Test the model with a new sequence
    test_sequence = np.array([7, 8, 9]).reshape((1, 3, 1))  # Reshape to match input shape
    prediction = model.predict(test_sequence)
    print("Prediction:", prediction)

simple_sequence_prediction()
```

**LSTM IMPLEMENTATION**

```python
from keras.datasets import imdb
from tensorflow.keras.preprocessing.text import Tokenizer
from keras.utils import pad_sequences
from keras import Sequential
from keras.layers import Dense,Embedding ,LSTM,Dropout ,Flatten
```

```python
(X_train,y_train),(X_test,y_test)=imdb.load_data()
```

```python
#text preprossing
X_train=pad_sequences(X_train,padding='post',maxlen=50)
X_test=pad_sequences(X_test,padding='post',maxlen=50)
```

```python
X_train.shape
```

```python
X_train[0]
```

```python
import tensorflow as tf
inputs=tf.keras.Input(shape=(50,))
x=tf.keras.layers.Embedding(10000,output_dim=1,input_length=50)(inputs)
x=tf.keras.layers.LSTM(32,return_sequences=False)(x)
outputs=tf.keras.layers.Dense(1,activation='sigmoid')(x)
model=tf.keras.Model(inputs=inputs,outputs=outputs)
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['acc'])
history=model.fit(X_train,y_train,epochs=5,validation_data=(X_test,y_test))
```

```python
import numpy as np
#Get the index of the test review you want to display
index=0 # change this to the desiredd index

#Get the actual review text
#since imdb.load_data() return tokenied data
#you need to reverse the tokenization to get the original text.

word_index=imdb.get_word_index()
reverse_word_index=dict([(value,key) for (key,value) in word_index.items()])
review=' '.join([reverse_word_index.get(i -3 ,'?') for i in X_test[index]])

#Get the actual label
actual_label=y_test[index]
#Get the model's prediction
prediction=model.predict(X_test[index].reshape(1,-1))

#print the result
print("Review:",review)
print("Actual label:",actual_label)
print("Model Prediction: ",prediction[0][0])
print("Predicted Label:",(prediction[0][0]>0.5).astype(int)) #1 if prediction >0.5 ,else 0
```
