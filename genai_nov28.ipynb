**Fine tunning a model**

1. adjusting a pre-trained model for a specific task or dataset.
2. pre trained model: algorithem developed on extensive data to perform a task.
3. learn new task or domain.
4. reduced time and computation.

How to fine-tune a model?
1. identify the model based on task.
2. prepare the data
3. build trainer
4. train a model

```python
from transformers import AutoModelForSequenceClassification

model_name="distilbert/distilbert-base-uncased-finetuned-sst-2-english"

model=AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
#load the tokenizer
from transformers import AutoTokenizer
tokenizer=AutoTokenizer.from_pretrained(model_name)
```

```python
tokenizer
```

```python
from datasets import load_dataset
dataset=load_dataset("stanfordnlp/imdb")
```

```python
dataset
```

```python
#use tokenizer in text
dataset=dataset.map(lambda row: tokenizer(row["text"],padding=True,max_length=512,truncation=True),keep_in_memory=True)
dataset
```

```python
dataset['train']
```

```python
dataset['train']['text'][1]
```

```python
dataset['train']['attention_mask'][1]
```

```python
from transformers import (
    Trainer,
    TrainingArguments
)
```

```python
#creaate training arguments
training_args=TrainingArguments(output_dir="./results")
#datacollectorwithpadding is a utility provided by hugging face tht ensure all data has same length during training

from transformers import DataCollatorWithPadding

trainer=Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset["train"],
    data_collator=DataCollatorWithPadding(tokenizer=tokenizer)
)

trainer.train()
```

```python
pipeline(task="sentiment-analysis",model="./fine_tuned_model")
text_example=" I am huge fan of comedies"
classifier(text=text_example)
```
