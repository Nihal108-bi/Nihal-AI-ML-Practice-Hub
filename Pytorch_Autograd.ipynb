```python
import torch

#inputs
x=torch.tensor(6.7)
y=torch.tensor(0.0)  #true table (binary)

w=torch.tensor(1.0)
b=torch.tensor(0.0)
```

```python
#Binary Cross-Entropy Loss for scaler
def binary_cross_entropy_loss(prediction,target):
  epsilon=1e-8 #to prevant log(0)
  prediction=torch.clamp(prediction,epsilon,1-epsilon)
  return -(target * torch.log(prediction) + (1-target)*torch.log(1-prediction))
```

```python
#Forwar pass
z=w*x+b
y_pred=torch.sigmoid(z)

loss=binary_cross_entropy_loss(y_pred,y)
```

```python
loss
```

```python
#derivalives
#1. loss with respect to the prediction(y_pred)
dloss_dy_pred=(y_pred-y)/(y_pred * (1-y_pred))

#2. prediction (y_pred) with repsect to z (sigmoid derivative)
dy_pred_dz=y_pred * (1-y_pred)
#3. z with repect to w and b

dz_dw=x
dz_db=1 #bias contributes diracty to z

dL_dw=dloss_dy_pred * dy_pred_dz * dz_dw
dL_db=dloss_dy_pred * dy_pred_dz * dz_db
```

```python
print(f"Maual Gradient of loss w.r.t weight (dw): {dL_dw}")
print(f"Manual Gradient of loss w.r.t bias (db): {dL_db}")
```

```python
x=torch.tensor(6.7)
y=torch.tensor(0.0)

w=torch.tensor(1.0,requires_grad=True)
b=torch.tensor(0.0,requires_grad=True)
```

```python
w
```

```python
b
```

```python
z=w*x+b
z
```

```python
y_pred=torch.sigmoid(z)
y_pred
```

```python
loss=binary_cross_entropy_loss(y_pred,y)
loss
```

```python
loss.backward()
print(f"Gradient of loss w.r.t weight (dw): {w.grad}")
print(f"Gradient of loss w.r.t bias (db): {b.grad}")
```
